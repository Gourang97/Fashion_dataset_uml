```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import glob
from scipy.misc import imread
from keras.preprocessing.image import load_img,img_to_array
from keras.layers import Flatten
from glob import iglob
from sklearn.decomposition import PCA
```

```python
img = load_img("C:/Users/goura/Google Drive/UML/HW/H4+H5/H4+H5/3096_colorPlane.jpg")
plt.imshow(img)
img_data = img_to_array(img)
x, y, _ = img_data.shape

# Normalizing RGB values
img_data = img_data/255

img_data.shape
```

```python
x_norm, y_norm = [], []
for i in range(x):
    x_norm.append(i/x)
for j in range(y):
    y_norm.append(j/y)
    
img_data.shape[0]
```

```python
new_img = np.zeros(shape = (img_data.shape[0], img_data.shape[1], 5))
x_norm = np.array(x_norm).reshape(len(x_norm),1)
x_norm_arr = np.tile(x_norm,481)
x_norm_arr.shape

y_norm = np.array(y_norm).reshape(len(y_norm),1)
y_norm_arr = np.tile(y_norm,321)
```

```python
new_img[:,:,0:3] = img_data
new_img[:,:,3] = x_norm_arr
new_img[:,:,4] = y_norm_arr.T
```
